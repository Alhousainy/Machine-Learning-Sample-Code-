```python
#First Read My Data 
path='D:\\Programming\\Machine Learning Projects\\My Projects\\Data lreg (one_variable).txt'
data=pd.read_csv(path,header=None , names=['population', 'profits'])
#Gradient Descent Equation Because To Solve Problem
#1- addin a new column called ones before the data
data.insert(0,'Ones',1)
print('\t The New Data Is : \n',data.head(10))
print('*'*40)
#==========================================================================================
#Seperate x(Training Of Data) from y (Target Variable)
cols=data.shape[1]
xinput=data.iloc[ : , 0 : cols-1]
youtput=data.iloc[ : , cols-1 : cols]
print('Input Data (X) Is \n ',xinput.head(10))
print('*'*40)
print('Output Data (Y) Is : \n',youtput.head(10))
```

```python
#Convert From Data Frames To Numpy Matrices 
xm=np.matrix(xinput.values)
ym=np.matrix(youtput.values)
theta=np.matrix(np.array([0,0]))
print('X Matrix Is \n',xm)
print('The Sahpe Of X Matrix Is ==> ',xm.shape)
print('*'*50)
print('The Theta Is ==> ',theta)
print('The Theta Shape Is ==> ',theta.shape)
print('*'*50)
print('The Y Matrix Is \n',ym)
print('The Shape Of Y Matrix Is ==> ',ym.shape)
```

```python
#To Calculate Cost Function 
def compute_cost(x,y,theta):
    cost = np.power(((x * theta.T)- y ), 2)
    return np.sum(cost)/ (2 * len (x))
print('The Cost Function (XM ,YM , Theta ) Is ==> ',compute_cost(xm,ym,theta))
print('**'*40)
#============================================================================================
#To Calculate The Gradient Descent 
def gradient_descent(x,y,theta,alpha,iters):
    temp =np.matrix(np.zeros(theta.shape)) # Make Temp Matrix Contain Same Boundaries In Theta [0. 0.]
    parameters=int(theta.ravel().shape[1])
    cost=np.zeros(iters)
    for i in range (iters):
        error = (x * theta.T)- y
        for j in range(parameters):
            term=np.multiply(error,x[:,j])
            temp[0,j]=theta[0,j]-((alpha/len(x)) * np.sum(term))
        theta = temp 
        cost[i] = compute_cost(x,y,theta)
    return theta , cost 
#======================================================================================================
alpha =0.001
iters =10000
g,cost=gradient_descent(xm,ym,theta,alpha,iters)
print('The Value Of Theta 1 & Theta 2 Is ==> ',g)
print('The Cost Function For 50 Values Is ==> ',cost[:200])
print('*'*50)
print('The Final (The Compute Cost Is Equal ) ==> ',compute_cost(xm,ym,g))
```

```python
#To Ready Data Becuase Draw The Best Fit Line
data1=np.linspace(data.population.min(),data.population.max(),200)
print('The Data After Divided To 200 Equal Part Is ==> \n',data1)
#To Make Hypothesis Equation ==> h(Theta)= theta 0 + theta 1 * x 
f=g[0,0]+(g[0,1] * data1)
print('**'*50)
print('The Hypothesis Equation Result Is ==> \n',f)
```

```python
#To Draw The Best Fit Line 
plt.style.use('ggplot')
fig,ax=plt.subplots(figsize=(5,5))
ax.set(title='Predicted Profits vs. Population Size ',xlabel='Populations',ylabel='Profits')
ax.plot(data1,f,color='green',linestyle='dotted',marker='*',alpha=0.5,label='Prediction')
ax.scatter(data.population,data.profits,color='red',label='Training Data')
ax.legend(loc=2)
```
